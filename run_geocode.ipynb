```python
from geopy.geocoders import Nominatim
import time
from random import randint
from geocoder import geocode
import pandas as pd
import numpy as np

user_agent = 'user_me_{}'.format(randint(10000,99999))
geolocator = Nominatim(user_agent=user_agent)

# Read CSV file into a dataframe
torino_houses = pd.read_csv('torino-listings-clean.csv')

# Initialize empty longitude and latitude columns
torino_houses['Longitude'] = np.NaN
torino_houses['Latitude'] = np.NaN

# Loop through addresses in dataframe to geocode each listing
for address in torino_houses['Address description']:
    location = geocode(geolocator, address, 1)
    
    if location is None: continue
    
    # Extract the longitude and latitude of current address
    torino_houses['Longitude'][address] = location.longitude
    torino_houses['Latitude'][address] = location.latitude
    
    # Maintain at least 1 second per request as per Nomatim policy, and random interval to avoid time-out
    time.sleep(randint(1,10))
    
torino_houses.to_csv('torino-listings-clean-latlong.csv', index=False)
    
    
    
```
